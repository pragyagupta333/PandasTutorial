# Creating Plots In Pandas

# Visual Check Of Data

```python
data.plot()
```

# Plot Only Calories Data

```python
data['Calories'].plot()
```

# Scatter Plot 

#### Create Scatter Plot with Duration As X-axis And Calories as Y-axis

The kind argument specifies type of plot you want to create.

```python
data.plot(kind = 'scatter', x = 'Duration', y = 'Calories')
```

So, We can conclude that higher the duration,more the calories burned

# Histogram 

#### A histogram needs only one column.

#### Create a Histogram Which Shows : How Many Workouts Burned Calories Between 250 to 500 units

```python
data["Calories"].plot(kind = 'hist')
```

The Histogram Tells That Over 100 Workouts Burned Calories Between 250 to 500 units

# Show each of the columns in a separate subplot.

Separate subplots for each of the data columns are supported by the subplots argument of the plot functions.

```python
data.plot(subplots = True)
```

# Plot Area Graph For DataFrame

```python
data.plot.area()
```

# Set label/name for axis and save the graph figure

```python
fig, axs = plt.subplots(figsize=(12, 4))        # Create an empty Matplotlib Figure and Axes
data.plot.area(ax=axs)                   # Use pandas to put the area plot on the prepared Figure/Axes
axs.set_ylabel("y-axis-label")          # Do any Matplotlib customization you like
fig.savefig("Graph.png")           # Save the Figure/Axes using the existing Matplotlib method.
plt.show()                                      # Display the plot
```

## What Is The Ratio Of Calories Burned To Duration Of Workout? Store Result In Separate Column

```python
data['Calories_duration_ratio'] = data['Calories']/data['Duration']
```

# Rename Column 'Calories_duration_ratio' TO 'Ratio_Calories_duration'
