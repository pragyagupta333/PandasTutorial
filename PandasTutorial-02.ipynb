### Fetch Passengers older Than 35 Years

To select rows based on a conditional expression, use a condition inside the selection brackets []

```python
above_35 = data[data['Age']>35]
```

```python
above_35.head()
```

### Fetch Passengers From Cabin Class 2 And 3.

The isin() conditional function returns a True for each row the values are in the provided list. 

```python
class_2_3 = data[data['Pclass'].isin([2,3])]
```

```python
class_2_3.head(3)
```

                                                OR

```python
class_2_3_ = data[(data['Pclass'] == 2) | (data['Pclass'] == 3) ]
```

```python
class_2_3_.head(3)
```

### Fetch Passenger Data For Which The Age Is Known i.e Not Null

The notna() conditional function returns a True for each row the values are not a Null value.

```python
age_not_null  = data[data['Age'].notna()]
```

```python
age_not_null.head()
```

## Fetch Names of the passengers older than 35 years.

data[data['Age']>35,'Name']

Above code won't work.

Here, we want a subset of both rows and columns i.e age rows>35 and column = name. In such cases use .loc

```python
data.loc[data['Age']>35,'Name']
```

When using loc/iloc, the part before the comma is the rows you want, and the part after the comma is the columns you want to select.

loc  : column names, row labels or a condition expression

iloc : position of columns

### Fetch Rows 10 To 25 And Columns 3 to 5.

Note : Rows and columns Starts from 0

```python
data.iloc[9:25 , 2:5]
```
